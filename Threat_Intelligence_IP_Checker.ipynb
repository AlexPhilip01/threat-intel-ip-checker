```python
import re
import requests
```

```python
log_data = """
Failed login from 192.168.1.50
Failed login from 45.33.32.156
Connection attempt from 8.8.8.8
Suspicious traffic from 185.220.101.1
"""

with open("network.log", "w") as f:
    f.write(log_data)

print("Sample log file created")
```

```python
def extract_ips(file_path):

    with open(file_path, "r") as f:
        data = f.read()

    ip_pattern = r'\d+\.\d+\.\d+\.\d+'
    ips = re.findall(ip_pattern, data)

    return list(set(ips))
```

```python
API_KEY = "ade949413dfe85cc0ac1e12d464cef1adab12be40989724aeca124ed39dfad9a7f3cc09f9002c3d8"

def check_ip(ip):

    url = "https://api.abuseipdb.com/api/v2/check"

    querystring = {
        'ipAddress': ip,
        'maxAgeInDays': '90'
    }

    headers = {
        'Accept': 'application/json',
        'Key': API_KEY
    }

    response = requests.get(url, headers=headers, params=querystring)

    data = response.json()

    score = data["data"]["abuseConfidenceScore"]

    print(f"\nIP: {ip}")
    print(f"Abuse Confidence Score: {score}")

    if score > 50:
        print("⚠️ Potentially Malicious IP")
    else:
        print("Low Risk")
```

```python
ips = extract_ips("network.log")

print("IPs found in logs:", ips)

for ip in ips:
    check_ip(ip)
```
